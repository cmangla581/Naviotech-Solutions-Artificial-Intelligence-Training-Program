                                                    Filtering Columns

To filter columns, from Pandas DataFrame, yoou can use the Filter method. The list of columns that you want to filter is passed to the filter method.

                                                  Concatenating Data 

Sometimes, we need to concatenate or join the multiple datas by Pandas Dataframes.  We create teh 2 dataframes suing the Pandas dataFrames using the Titanic Data. 

```python
titanic_pclass1_data = titanic_data[titanic_data.Pclass == 1]
print(titanic_pclass1_data.shape) 
print(titanic_pclass1_data)

titanic_pclass2_data = titanic_data[titanic_data.Pclass == 2] 
print(titanic_pclass2_data.shape)
print(titanic_pclass2_data)
```

There are two ways to concatenate datasets horizontally. We can call the append() method via the first dataframe and pass the second dataframe as a aparameter to the append() method.  

```python
final_data = titanic_pclass1_data.append(titanic_pclass2_data, ignore_index=True) 

print(final_data.shape) 

# Here it will show aeeror as the dataframe.append() version has been removed and now e hae to use the concat method  
```

```python
final_data = pd.concat([titanic_pclass1_data, titanic_pclass2_data]) 

final_data

```

                                                    Sorting Dataframes

To dort the pnadas datframes, we can use the sort_values function of Pandas dataFrame. The list of columns used for sorting needs to be passed by the attribute of sort values() method.  

```python
age_sorted_data = titanic_data.sort_values(by = ["Age"]) 

age_sorted_data.head() 
```

```python
# To sort values by the descending ordr, we need to pass False as value for the ascending attribute of the sort_values. 

age_sorted_data = titanic_data.sort_values(by = ["Age"], ascending = False) 

age_sorted_data.head()
```

```python
# In this code , the age and fare are sorted in the descending order 

age_sorted_data = titanic_data.sort_values(by = ["Age", "Fare"], ascending = False) 

age_sorted_data.head()
```

                                                   Apply Function 

The apply() function is used to apply a function on multiple rows or on rows of a particular function.  A lambda function is passed to apply function. It specifies the function performed by apply function.  

The apply() function adds 2 to all the values in the Pclass column oof titanic dataset. 

```python
updated_class = titanic_data.Pclass.apply(lambda x : x + 2)
updated_class.head()
```

```python
def mult(x): 
    return x*2 

updated_class = titanic_data.Pclass.apply(mult)
updated_class.head()

# All the values in the Pclass will be multiplied by 2 
```

                                                    Arithmetic Operators 

The where clause from the numpy library is used to perform  the arithmetic operation of pandas dataframe. The where clause is used to add 5 to the rows in the fare column. where passengers ages are greater than 20 

```python
import numpy as np 
titanic_data.Fare = np.where(titanic_data.Age > 20, titanic_data.Fare + 5, titanic_data.Fare) 

titanic_data.head()
```
