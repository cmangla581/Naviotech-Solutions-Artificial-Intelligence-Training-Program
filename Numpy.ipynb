Reshaping NumPy arrays : A NumPy array can be reshaped using the reshape() function. Its important to mention the product of the rows and columns in the reshaped array must be equal to the product of the rows and columns in the original array. 

```python
uniform_random = np.random.rand(4,6)
uniform_random2 = uniform_random.reshape(3,8)

print(uniform_random)

print(uniform_random2)
```

Array Indexing and Slicing :  NumPy arrays can be sliced and indexed. Slicing an array means dividing an array into mutiple parts. 
NumPy arrays are indexed into multiple parts. NumPy arrays are indexed just like the normal lists. Indexes in NumPy arrays start from 0 which means that the first itemof a NumPy array is stored at the 0th index. 

```python
row1 = [10,12,13]
row2 = [45,32,16]
row3 = [45,32,16]

nums_2d = np.array([row1, row2, row3])
print(nums_2d[:2 , :])
```

```python
row1 = [10,12,13]
row2 = [45,32,16]
row3 = [45,32,16]

nums_2d = np.array([row1, row2, row3])
print(nums_2d[:, :2])
```

```python
import numpy as np 
row1 = [10,12,13]
row2 = [45,32,16]
row3 = [45,32,16]

nums_2d = np.array([row1, row2, row3])
print(nums_2d[1:, 1:])

# Here we have done such a slicing that there are last 2 rows and last 2 columns 
```

                                          NumPy for Linear Algebra Operations 
Data Science makes extensive use of the linear algebra. The suppot for performing the advanced linear algebra functions in a fast and eficient way.  

Hence it makes NumPy the most routinely used library in Python to be used.  

```python
A = np.random.randn(4,5) 
B = np.random.randn(5,4) 

Z= np.dot(A,B) 

print(Z)
```

```python
row1 = [1,2,3] 
row2 = [4,5,6] 
row3 = [7,8,9] 

nums_2d = np.array([row1, row2, row3]) 

multiply = np.multiply(nums_2d, nums_2d)  

print(multiply)
```

```python
row1 = [1,2,3] 
row2 = [4,5,6] 
row3 = [7,8,9] 

nums_2d = np.array([row1, row2, row3])  

inverse = np.linalg.inv(nums_2d)  

print(inverse) 

# Here since the detremonatn = 0 , so its showing error hence no inverse of the matrix 
```

```python
import numpy as np 
row1 = [10,12,13]
row2 = [45,32,16]
row3 = [34,21,89]

nums_2d = np.array([row1, row2, row3]) 

inverse = np.linalg.inv(nums_2d) 

print(inverse)
```

```python
# Finding the matrix determinant 

import numpy as np 
row1 = [10,12,13]
row2 = [45,32,16]
row3 = [34,21,89]

nums_2d = np.array([row1, row2, row3]) 

determinant = np.linalg.det(nums_2d) 

print(determinant )
```

```python
# Finding the trace of a matrix  
# Trace refers to the finding of the sum of all elements along the diagonal of a matrix 

row1 = [1,2,3] 
row2 = [4,5,6]
row3 = [7,8,9] 

nums_2d = np.array([row1, row2, row3])

trace = np.trace(nums_2d) 
print(trace) 
```
