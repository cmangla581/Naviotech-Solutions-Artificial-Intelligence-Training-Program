Artificial Intelligence is a branch of CS that is concerned with building smart and intelligent machines. 

AI enables computer and machines to simulate human learning, comprehension ans problem solving. 

Machine Learning is a subset of Artificial Intelligence and Deep Learning is a subset of Machine Learning. 

ML is a technique to implement AI that can learn from data by themselves without being programmed. 

Deep Learning is a subset of Machine Learning which uses the Artificial Neural Networks to learn from large datasets. 
The concept of Neural Networks resembles the Human Neurons which perform the function of signalling the messages across the nerves and perform the function of response to stimuli.   

Hence Deep Learning technique is useful  in handling large datasets with more accuracy.

                                               Supervised Machine Learning 

The Supervised machine Learning is the type of machine learning that learns from labelled data set. The main types of Supervised Machine Learning include : 

1. Classification  
2. Regression  

Classification :  
Classification is the type of supervised machine learning which is about predicting a class or discrete values.  It basically involves calculating a categorical outcome. 
For example, it calculates the outcomes like Male or Female, Yes or No, True or False.  

The different algorithms for the  classification Supervised Machine Learning include the : 
1. Decision Tree
2. Random Forest Classifier
3. K Nearest Neighbours 
4. Logistic Regression 

Regression : 
Regression is the the type of supervised machine learn99ing which is about predicting a numerical outcome. 
Example : It predicts a numerical value as the outcome. 

The different algorithms for the Regression Supervised MAchine Learning include the : 
1. Linear Regression'
2. Ridge Regression 
3. Lasso Regression

                                                        Deep Learning

Its a sub field of machine learning that uses the Artificial Neural Networks to learn from large  datasets.

It takes inspiration from the working mechanism of the Neurons of the Brain. 

The basic difference between the Machine Learning and the Deep Learning is that the :  

In Machine Learning, the process takes place as the : 
Input >>>> Feature Extraction >>> Classification/Regression >>>> Output 

In Deep Learning, the process takes place as the : 
Input >>>>>>   Artificial Neural Networks >>>> Output 

The Artificial Neural Networks include the Feature Extraction alongwith Classification and Regression on large datasets.
