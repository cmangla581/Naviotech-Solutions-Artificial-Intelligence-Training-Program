Conditional Statements :  

Conditional Statements in Python are used to implement conditionallogic in Python. Conditional statements help you decide whether to execute a certain code block or not. 

There are 3 main types of conditional statements : 
1. If
2. Else 
3. if - elif 

```python
# If statement  
if 10 > 5 : 
    print("10 is greater than 5")
```

```python
# if else statement  

if 5 > 10 : 
    print("5 is greater than 10")

else : 
    print("10 is greater than 5")
```

```python
# if elif and else statements  

if 5 > 10 : 
    print("5 is greaater than 10")

elif 8 < 4 : 
    print("8 is smaller than 4")

else : 
    print("5 is not greater than 10 and 8 is not smaller than 4")
```

Iteration Statements : Iteration statements are also known loops used to iteratively executa a certain piece of code.  

They are 2 main types of iteration statements in Python : 
a. For Loop 
b. While Loop 

```python
# example of loops 
items = range(5) 

for item in items : 
    print(item)
```

```python
c = 0 
while c < 10 : 
    print(c)
    c += 1
```

                                                 Objects and Classes

Python supports the Object Oriented Programming in Python. In OOP, any entity can perform some function and have some attributes is impemented in the form of an object. 

```python
# Example of the code 
class Fruit : 
    name = "apple"
    price  = 10 

    def eat_fruits(self): 
        print("Fruits has been eaten") 

f = Fruit()
f.eat_fruits()
print(f.name)
print(f.price)
```

```python
class Fruit : 
    name = "apple"
    price = 10 

    def __init__(self, fruit_name, fruit_price): 
        Fruit.name = fruit_name 
        Fruit.price = fruit_price 

    def eat_fruit(self): 
        print("Fruit has been eaten") 


f = Fruit("Orange", 15)
f.eat_fruit()
print(f.name)
print(f.price)
```
