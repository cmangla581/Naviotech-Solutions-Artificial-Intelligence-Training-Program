Machine Learning is a branch of artificial intelligence that enables the cmputer programs to automatically learn and improve from the experience.  Machine Learning algorithm learn from datasets and then based on patterns identified from the dataasets, make predictions on the unseen data. 

Machine Learning Algorithms can be classified into 2 algorithms : 
1. Supervised Machine Learning Algorithm 
2.  Unsupervised Machine Learning Algorithm 

Supervised Machine Learning Algorithms are those algorithms where the input dataset and the corresponding output or the true predictions are available and the algorithms try to find the relationship between the inputs and outputs. 

In Unsupervise Machine Learning, the true labels for the outputs are not known. The algorithms try to find the similar patterns in the data.  The algorithms try to find the similar patterns in the data. Clustering algorithms are a typical example for the unsupervised learning. 

Supervised Machine Learning Algorithms are further divided in 2 types : 
1. Regression Algorithms 
2. Classification Algorithms 

Regression Algorithms predict a continuous value. For example, the price of a house, blood pressure of a person, and the student's score in a exam. 

Classification Algorithms on the other hand predict a  discrete value such as whether or not, tumor malignant or not or a student is going to pass or fail. 

                                            Preparing Data for Regression Problems

Machine Learning algorithms require the data to be in certain format before the algorithms can be trained on the data. We see the various data pre processing step that we need to perform before we can train the machine learning algorithms by Sklearn library. 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
sns.get_dataset_names()

# These are all the datasets present in the seaborn library 
```

```python
# Loading the tips dataset
tips_df = sns.load_dataset("tips") 
tips_df.head()
```

                                       Dividing the Data into Features and Labels 
We device the data into features and labels. Our label sets consist of the values from the tip column. while the features set consist of the values from the remaining columns. 

```python
x = tips_df.drop(["tip"], axis = 1) 
y = tips_df["tip"]
```

```python
# printing the fatures dataset 
x.head() 
```

```python
y.head()
```

                                        Converting Categorical Data to Numbers
Machine Learning algorithms for the msot part can only work with numbers. Hence for those cases its importatn to convert the categorical datas into numeric format.  

The first step is to create the dataset of the numeric values. To do so , we drop the categorical columns from the dataset. 

```python
numerical = x.drop(['sex', 'smoker', 'day', 'time'], axis = 1)
numerical.head()
# Here the categorical outcome is seperated from the numerrical datas 
```

```python
categorical = x.filter(['sex', 'smoker', 'day', 'time']) 
categorical.head() 
```

One of the most common approaches to convert a categorical outcome into numeric one is via one hot encoding. In one hot encoding, every unique value in the original columns, a new column is created.  

For example in sex column, two new columns of Male and Female are created.  If the original sex column contained male, 1 is addded in the newly created Male Column. While if the original column contained a Female, then 1 is added to the Female Column. 

However we do not really need  2 columns. A single column Female is enough since when a customer is female. we can add 1 in the Female Column else 1 can be added in that column. Hence, we need the N-1 one hot encoded columns for all the n values in the original columns.  

```python
import pandas as pd 
cat_numerical = pd.get_dummies(categorical, drop_first = True)
cat_numerical.head() 
```

```python
# Here this table is still in the form of the True or False but we need to convert it in the form of 0 and 1  
# 1 stands for True and 0  stands for False  

cat_numerical = pd.get_dummies(categorical, drop_first = True).astype(int) 
cat_numerical.head()
```

The final step is to join the numerical columns with the one hot encoded columns. To do so we have to use the concat() function from the Pandas Library. 

```python
x= pd.concat([numerical, cat_numerical], axis = 1) 
x.head() 

# Here we have concatenated the numerical and the categorical converted numerical data  
```

                                        Dividing the Data into Training and Tests Sets 

After we train a machine Learning algorithm, we need to evaluate it to see that how well it performs on unseen data.  

Therefore, we divide the data into two swts:  Training Set and Testing Set. The dataset is trained via the training set and the evaluated test set.  

To split the function into train_test_split() function from the Sklearn library.  

```python
#  The following script divides the data into an 80 percent training set and the 20 percnt test data 

from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = 0.20, random_state = 0)
```

                                              Data Scaling and Normalization 

The final step before the data is passed to amchine learning algorithms is to scale the data. Some columns of the dataset contain small values. it is betterr to convert all values to a uniform scale. 

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler() 

# scaling the training set 
x_train = sc.fit_transform(x_train) 
# Scaling the test set 
x_test = sc.transform(x_test)
```

We have to convert the data into a format that can be used to train the machines learning algorithms for regression from the Sklearn Library. The details including the functionalities and usage of all the machine learning algorithms. 

                                               Linear Regression 

Linear Regression is a linear model that presumes a linear relationship between the inputs and outputs and minimizes the cost of error between the predicted and the actual outputs using the functions like the mean absolute error between teh different data points. 

                                             Why use the  Linear Regression ? 
The Linear Regression Algorithm is particularl useful because :  
1. Linear Regression is simple to implement and easily interpretable algorithm. 
2. Takes less training time to train for huge datasets 
3. Linear Regression Coefficients are easy to interpret. 

                                            Disadvantages of Linear Regression 
the followingare the disadvantages of Linear Regression which are as follows : 
1. Performance is easily affected by outlier presence. 
2. Assumes a Linear Relationship between the dependent and the independent variables which can result in an increased erroor. 

                                          Implementing Linear Regression with Sklearn
To  implement the Linear Regression with the sklearn, we can use the Linear Regression from the sklearn.linear_model module. 
To train the algorithm, the training and the tests sets like the X_train and the X_test. 

The test set is passed to the predict() method of the class to make predictions. 

```python
# The code for training and making the predictions with liner regression algorithm .
from sklearn.linear_model import LinearRegression 

# training the algorithm '
model = LinearRegression() 
regressor  = model.fit(x_train,  y_train) 

# making predictions on the test set 
regressor 
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

Once we have trained a model and have made predictions on the test set, the next step is to know, how well yur model has performed for making predictions on the unknown set. 

                                                    Mean Absolute Error 
Mean Absolute error is calculated by taking the average of the absolute error obtained by subtracting the real values from the predicted values.  

It is the average of the subtraction of the real values from the predicted values. 

                                                          Mean Squared Error 
Mean Squared Error (MSE) is similar to MAE. The error for each record is aquared.  So basically in case of mean squared errors, we just square the difference of the predicted and real values and the average is taken out. 

So this is to calculate a huge difference between the predicted and actual values.  

                                            Root Mean Squared Error
The root mean squared error is the square root of the mean squared error. 

```python
from sklearn import metrics 

print("Mean Absolute error: ", metrics.mean_absolute_error(y_test, y_pred)) 
print("Mean Squared Error ", metrics.mean_squared_error(y_test, y_pred)) 
print("Root Mean Squared Error: ", np.sqrt(metrics.mean_squared_error(y_test, y_pred))) 
```

                                                    KNN Regression 

KNN stands for the K nearest neighbors. KNN is lazy learning which is based on finding the Euclidian distance between the different data points. 

                                               Why use KNN Algorithm ? 
1.  KNN Algorithm doesn't assume theany relationship between the features. 
2.  Useful for a dataset between the data localization is important. 
3. Only have to tune the parameter K, which is the number of nearest neighbours. 
4.  no training is needed as it is a lazy learning algorithm. 
5. Recommender Systems and finding  semantic similarity between the documents are a major applications of the KNN algorithm. 

                                              Disadvantages of KNN Algorithm 
The following are the disadvantages of KNN algorithm which include : 
1. You have to find the optimal value of K. 
2. not suitable for very high dimensional data. 

                                               Import KNN Algorithm with Sklearn 
With Sklearn, it is extremely easy to implement the KNN Regression we use the KNeighborsRegressor class. The process of training and testing, is same as linear regression.  

For training, we need to call the fit() method and for testing we need to call the predict() method. 

```python
# The sript for the training, testing and the evaluating the KNN regression for predicting the values for the tip column 

from sklearn.neighbors import KNeighborsRegressor 
knn_reg = KNeighborsRegressor(n_neighbors = 5) 

regressor = knn_reg.fit(x_train, y_train) 
regressor 
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
from sklearn import metrics 

print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_pred)) 

print("Mean Squared error: ", metrics.mean_squared_error(y_test, y_pred)) 

print("Root Mean squared Error; ", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

                                            Random Forest Regression 
Random Forest is a tree based algorithm that converts features into tree nodes and then uses the entropy loss to make predictions. 

                                                Why use Random Forest Algorithms ? 
Randm Forest Classifier is useful when : 
1. Ypu have lots of missing data or imbalanced data. 

2. With large number of trees, we can avoid overfitting while training. Overfitting occurs when machine learning moodels perform better on the training set. 

3. The random forest algorithm can be used when you have very high dimensional data. 
4. through cross validation, the random forest algorithm can return higher accuracy. 
5. The random forest algorithm can solve both classificationand regression tasksand find its application in a variety of tasksranging from credit card fraud to stock market prediction. 

                                            Disadvantages of Random Forest Algorithms 
There are 2 major disadvantages of the Random Forest Algorithms : 
1. Using a large number of trees can slow down the algorithm. 
2. Random Forest Algorithm is a predictive algorithm which can only predict the future and  cannot explanin what happended in the past using the dataset. 

                                      Implementing Random Forest Regressor using Sklearn 
random Forest Regressor Class from Sklearn ensemble mdule can be used to implement random forest regressor algorithms 

```python
# training andtesting the random forest 
from sklearn.ensemble import RandomForestRegressor 
model = RandomForestRegressor(random_state = 42, n_estimators = 500) 

regressor = model.fit(x_train, y_train) 
regressor 
```

```python
y_pred = regressor.predict(x_test) 
y_pred 
```

```python
from sklearn import metrics 
print('Mean Absolute error: ', metrics.mean_absolute_error(y_test, y_pred)) 

print("Mean Squared error: ", metrics.mean_squared_error(y_test, y_pred)) 

print("Root Mean Squared Error ", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
      
```

                                            Support Vector Regression 
The support vector machine is classification as well as regression algorithms which minimizes the error between the actual predictions and predicted predictions by maximizing the distance between the hyperplanes that contain the data for various records.  

                                                    Why use SVR Algorithms ? 
The support vector machine is a support vector machine variant for regression. SVm has the following usages : 
1. it can be used to perform the regression or classification with high dimensional data. 
2. With kernal trick, SVM is capable of applying the regression and classifcation to non linear datasets. 
3. SVm algorithms are used for ordinary classification or regression and thats why they are commonly used as ranking algorithms. 


                                                Disadvantages of SVR Algorithms 
There are major disadvantages of SVR algorithms : 
1. Lots of parameters  to be optimized in order to get the best performance. 
2. Training can take a long time on large datasets. '
3. it yields poor results if the number of features is greater than the number of records in a dataset. 

                                            Implementing SVR using Sklearn 
With the Sklearn library, we can use the SVM to implement support vector regression algorithms as shown below. 

```python
# training and testing the SVM  

from sklearn import svm 
svm_reg = svm.SVR() 

regressor = svm_reg.fit(x_train, y_train) 
regressor
```

```python
y_pred = regressor.predict(x_test) 
y_pred 
```

```python
from sklearn import metrics 

print("Mean Absolute error: ", metrics.mean_absolute_error(y_test, y_pred)) 

print("Mean Squared Error; ", metrics.mean_squared_error(y_test, y_pred)) 

print("root Mean Squared error: ", np.sqrt(metrics.mean_squared_error(y_test, y_pred))) 
```

                                                K Fold Cross Validation 
earlier we divided the data into the 80 percent training set and 20 percenttest set.  Only 20 percent of the data is used for testing and that 20 percent is never used for training. 

For more stable results, it is recommended that all the parts of the dataset are used atleat once for training and once for testing. 

the K fold cross validation can be used to do so. ith K fold cross validation, data is divided into 5 equal parts like K1, K2, K3, K4 and K5. 
in the first iteration, K1-K4 is used for training while K5 is used for testing. in the second test, k1, k2, K3 and K5 are used for training and K4 is used for testing.  

Each part is used at leat once for testing and once for training. We can use the cross_val-score() function from the sklearn model selection module to perform the cross validation. 

```python
from sklearn.model_selection import cross_val_score 

print(cross_val_score(regressor, x, y, cv=5, scoring = "neg_mean_absolute_error")) 

# the output shows the mean absolute value for each of the k folds. 
```
