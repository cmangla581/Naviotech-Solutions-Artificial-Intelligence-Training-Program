                                    Preparing Data for Classification Problems 

Like regression here also we have to first convert the data into a specific format before it can be used to train the classification algorithms. 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
```

The data used is of the Bank Customer Churn Prediction which basically has the record of the customers who left the bank six months after various information is recorded about them. 

```python
churn_df = pd.read_csv("Bank_Customer_Churn_Prediction.csv") 
churn_df  

# Here 1 stands for those who have exited and 0 for those who have not exited  
```

```python
churn_df.head() 
# Top 5 columns from the data 
```

The output shows the dtaset contain the information such as the  customer_id, geography, gender, age, etc.  
We do not need the RowNumber, Customerid, and surname Columns in pur dataset sice they do not help in predicting if a customer will churn or not. 

```python
churn_df = churn_df.drop(['customer_id'], axis = 1)  
churn_df
```

                                         Dividing Data into Features and Labels 
We have to divide the data into features and labels. The features set that is the Z in the following script contains all the columns except the Exited Columns. on the other hand, the labels set that is the y contains only the Exited columns. 

```python
x = churn_df.drop(['churn'], axis = 1) 
x  
# The churn column has beenn removed 
```

```python
y= churn_df['churn'] 
y 
# This is the churn column which shows that 
```

```python
x.head() 
```

```python
y.head()
```

                                    Converting the Categorical Data into the Numerical data 
For Classification problems too, we need to convert the categorical data to the nuerical data. This helps in easy implementation of the algorithm. 

```python
# Presenting the numerical data  
numerical = x.drop(['country', 'gender'], axis = 1) 
numerical.head() 

# Country and gender has been removed from the table dataset 
```

```python
# Presenting the categorical data  '
categorical = x.filter(['country', 'gender']) 
categorical.head() 
```

```python
# we have to use the one hot encoding approach to convert the categorica; data into the numerical data  
import pandas as pd 
cat_numerical = pd.get_dummies(categorical, drop_first= True).astype(int)
cat_numerical
```

```python
# Now we need to cancatenate the numerical and the categorical converted numerical data  in one tablle 
x= pd.concat([numerical, cat_numerical], axis = 1) 
x 
```

                                        Dividing the Data into Tranining and Test Splits 
Here, we divide the data into two sets, the training set and the testing set. The dataset is trained via the training set and evaluated on the test set.  

To split the data into train and test functions, we use the train_test-split() function from Sklearn library. There is 80 percent training set and 20 percent testing set. 

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.20 , random_state = 0) 
```

```python
x_train # data on which the model is trained 
```

```python
x_test # data on which model is tested
```

```python
y_train  # data on which model is trained 
```

```python
y_test # Data on which model is tested 
```

                                                   Data Scaling and Normalization 
The last step before the data is passed to machine learning algorithm is that the data is made to scale.    Beacuse some values in the data are too small and some are too large, so its necessary to scale the data to convert the data into a uniform scale. 

```python
# we use the function of Standard Scaler Preprocessing to scale the data  

from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()  

x_train = sc.fit_transform(x_train) 
x_test = sc.transform(x_test)  
```

So basically, by applying the above data, we have converted the dat into a step that can be used to train the machine learning algorithms for classification from the Sklearn library. 

                                                  Logistic Regression 

Logistic Regression is a linear model which makes the classification by passing the output of a linear regression by a sigmoid function. The process of training and making the predictions with logistic regression is as :  

```python
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression() 

classifier = model.fit(x_train, y_train) 
classifier 
```

```python
y_pred = classifier.predict(x_test) 
y_pred 
```

Once we have trained a model and have to make predictions based on the test set, we have to know how weel our model performs.  
There are various metrics to evaluate a classification method.  Some most commonly used calssification metrics are F1, recall, precision,  accuracy and confusion matrix. 

True Negative : True negatives are those values which were in real the false values and model also predicted them as the false values. 

True Positive : True positives are those values which are in real true values and the model also predicted them as the true values. 

False Negatives : False negatives are those which are actually true but are predicted false by the ml model. 

False Positive : Labels that are actually false but are prediccted true by the model are called false positives.  

The ordered and the structured way of writing the positives and negatives is known as the Confusion Matrix.  

                                                        Precision 
Another way to analyze a classification algorithm is by calculating the precision which is basically obtained by dividing the true positives by the sum of true positive and false positive.  

                         Precision = true Positive / (true positive + false positive)

                                                                Recall 
Recall is calculated by dividing the true positives by the sum of true positives and false negatives.  
             
                                 recall = true positive / (true positive + false negative)

                                                    F1 Measure 
F1 Measure is simply the harmonic mean of the preccision and the recall. It is calculated as :  
                                 
                                 F1 Measure = 2[(precesion * recall) / (precesion +  recall)] 

                                                         Accuracy 
Accuracy refers to the number of correctly predicted labels divided by the total number of observations in the dataset.  

                                             Accuracy = (tp + tn) / (tp + tn + fp + fn)

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score 

print("The confusion Matrix\n", confusion_matrix(y_test, y_pred)) 

print("The Classification Report: ", classification_report(y_test, y_pred)) 

print("The Accuracy Score: ", accuracy_score(y_test, y_pred)) 

# The record showws that the 81 percent of the  records in the test set, logistic regression correctly predicted whether a customer will leave the bank or no
```

                                                    KNN Classifier 
KNN stands for the K nearest neighbors.  KNN is a lazy algorithm which is based oon finding the Euclidean distance betweenn the different data points. 

The process of training and testing is same as logistic regression. For training, we need to call the fit() method and for testing we need to call the predict() method.  

the code for showing the process of training, testing and evaluating the KNN Classification is as : 

```python
from sklearn.neighbors import KNeighborsClassifier 

model = KNeighborsClassifier(n_neighbors = 5)  

classifier = model.fit(x_train, y_train) 
classifier
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score 

print("The Classificartion Report: ", classification_report(y_test, y_pred))

print("The Confusion matrix: ", confusion_matrix(y_test, y_pred)) 

print("The Accuracy Score: ", accuracy_score(y_test, y_pred)) 

# Here the accuracy iis around to be 0.827 which is around 82 percent 
```

                                                Random Forest Classifier 

Like the random Forest Regressor, the Random Forest Classifier is a tree based algorithm that converts the features into the tree nodes and then uses the entropy loss to make the classification predictions.  

The Random Forest Classifier csklearn.ensemble an be used to implement the random forest regressor algorithm in Python.  

In  Random forest we use the random_state and the n_estimators which are important parameters when configuring the Random Forest Classifer or even a Regressor. 

                                                        random_state  
The random_state parameter controla the randomness of the model's training process. 
In the random_stae, the randomness is used to select the data samples and features for each individual decision tree. 
It makes the result productive. 

                                                    n_estimators 
The n_estimators parametrs defines the number of decision trees in the forest.  A Random Forest is an ensemble learning method that builds many independent decision trees and merges their predictions to get a more accurate and stable result. 

More number of trees generally help in having a more robust and better performing model as the ensemble averages out individual tree errors more effectively. 

```python
from sklearn.ensemble import RandomForestClassifier 

model = RandomForestClassifier(random_state = 42, n_estimators = 500) 

classifier = model.fit(x_train, y_train) 

classifier
```

```python
y_pred = classifier.predict(x_test) 
y_pred 
```

```python
y_prob = classifier.predict_proba(x_test)[:,1]  
y_prob 
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score

print("The Confusion matrix: ", confusion_matrix(y_test, y_pred)) 

print("The classification Report; ", classification_report(y_test, y_pred)) 

print("The Accuracy Score: ", accuracy_score(y_test, y_pred)) 

print("ROC-AUC Score: ", roc_auc_score(y_test, y_prob))  

# Here the accuracy comes out to be 0.865 which is around 86 percent 
```
