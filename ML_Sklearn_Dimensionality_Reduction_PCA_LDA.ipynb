Dimensionality Reduction refers to the reducing the number of features in a dataset in such a way that the ovrall performance of the algorithms trained on the dataset is minimally affectedd. 

With the dimensionality reduction, the trainig time of the statistical algorithms gets reduced and also and data can be visualized more easily since it is not easy to visualize the datasets in higher dimensions. 

There are 2 main approaches for the Dimensionality Reduction : 
1. Pricipal Component Analysis (PCA) 
2. Linear Discriminant Analysis (LDA) 

                                        Principal Component Analysis                                                               

Principal Component Analysis is an unsupervised dimensionality reduction technique that doesn't depend on the labels of a dataset. PCA prioritizes features  on the basis of their ability to cause the maximum varince in the output. 

The idea behind the PCA is to capture the features that causes the maximum variance in the output is called the first principal component. The feature that causes the second highest highest variance is called the second princiapl component. 

Advantages of PCA : 
1. Correlated features can be detected and removed using PCA 
2. Reduces the overfittingbecause of reduction inn number of features 
3. Model training can be expedited.  

Disadvantages of PCA : 
1.  you need to standardize the data before you apply PCA. 
2. The independent variable becomes less integrable. 
3. Some amount of information is lost when you reduce features.  

                                        Implementing PCA with Sklearn Library

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
```

```python
# importing the dataset 
iris_df = sns.load_dataset("iris") 
iris_df.head() # here we are printing the dataset header 
```

```python
print(iris_df['species'].unique()) 
# Here as we can see that the unique features got printed 
```

The above output shows that the dataset contain four features: sepal_length, sepal_width, petal_length and one output label know  as the species.  For PCA we use only the feature set as it is an unsupervise learning. 

```python
x = iris_df.drop(['species'], axis = 1) 
y = iris_df.filter(['species'], axis = 1) 
```

```python
x  # The features et 
```

```python
y # The features label 
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 
y = le.fit_transform(y) 
y 
```

    Before we apply the PCA on a dataset, we will divide into the training set and test sets

```python
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x ,y, random_state = 42, test_size  =0.20) 
```

```python
x_train 
```

```python
x_test 
```

```python
y_train 
```

```python
y_test 
```

```python
# Applying the scaling on the training and test data 
from sklearn.preprocessing import StandardScaler 
sc= StandardScaler() 
x_train = sc.fit_transform(x_train) 
x_test = sc.transform(x_test) 
```

To apply PCA, we have to do is to import the PCA from the sklearn.decomposition module. Next is to apply the PCA to the training set, pass the training set to the to the fit_transform() method to the PCA class object. 

```python
# importing the PCA class 
from sklearn.decomposition import PCA 
pca  = PCA() 
pca 
```

```python
# traning the PCA on training data 
x_train = pca.fit_transform(x_train) 
x_train  
```

```python
# making prediction on test data 
x_test = pca.transform(x_test) 
x_test 
```

Once you have applied, PCA on a dataset, we can use the explained_variance_ratio_feature to print the variance caused by all features in the dataset. 

```python
# printing the variance ratios 
variance_ratios = pca.explained_variance_ratio_ 
print(variance_ratios) 
```

The output shows that the 72.55 percent of the variation is showed by the first component of the dataset. The 23.97 percent of variation is showed by the second princiapl component of the variation.  

Now, if we select the two princiapl components, we see that the total variation comes out to be 96.19 percent. 

To select the  2 pprincipal components, we we have to pass 2 as the vale to the n_omponents of the PCA Class. 

```python
# Use one principal commponent 
from sklearn.decomposition import PCA 
pca =  PCA(n_components = 2) 
pca 
```

```python
x_train = pca.fit_transform(x_train) 
x_test = pca.transform(x_test) 

variance_ratios = pca.explained_variance_ratio_ 
print(variance_ratios)  

# Here as we can see that we got 2 variance ratios 
```

Let us train a classification model using the logistic regression which predicts the label of iris plan tusing the two principal components of features, instead of original four features. 

```python
# making the predictions using the logistic regression 
from sklearn.linear_model import LogisticRegression 

model = LogisticRegression() 
classifier = model.fit(x_train, y_train) 
classifier 
```

```python
y_pred = classifier.predict(x_test) 
y_pred 
```

```python
from sklearn.metrics import accuracy_score 
print(accuracy_score(y_test, y_pred)) 

# The accuracy comes out to be around 90 percent  
```

The above output  shows that even with 2 features, the model accuracy is about 90 percent 

```python
import matplotlib.pyplot as plt 

plt.scatter(x_test[:,0], x_test[:,1], c = y_test, cmap = 'rainbow') 
```

                                                Linear Discriminant Analysis (LDA)

Linear Discriminant Analysis is a supervised dimensionality reduction technique where a decision boundary is fromed around the data points belonging to each cluster of a class. The data projections are projected to new dimensions in a way that the distance between the data points within a cluster. 

                                                                Why use LDA ? 

The following are the advantages of LDA : 
1. Reduces overfitting because of reduction in features. 
2. Model training can be expedited 

                                                        Disadvantages of LDA

There are 3 major disadvantages of LDA : 
1. Not able to correlated features
2. Cannot be used on  unlabelled data or unsupervised data 
3. Some amount of information is lost when you reduce features. 

                                              Implementing LDA with Sklearn Library

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
```

```python
# importing the dataset 
banknote_df = pd.read_csv("banknotes.csv") 
banknote_df 
```

```python
x = banknote_df.drop(["class"], axis = 1) 
y = banknote_df.filter(["class"]) 
```

```python
x 
```

```python
y 
```

Here also, we have to divide into the tranining and test splits. 

```python
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x ,y, test_size = 0.20, random_state = 42) 
```

Like PCA, here also we have to scale the data before the machine learning can be applied to it. 

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler() 

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test) 
```

To apply LDA via Sklearn, what we all need to do is to import the The Linear Discriminant Analysis class from teh Sklearn.decomposition module.  To apply LDA to the traning set, we have to pass the training set to the fit_transfoem() method of the LDA class object. 

```python
# Importing the LDA Class 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA 

# creating object of the LDA Class 
lda = LDA() 

lda 
```

```python
x_train = lda.fit_transform(x_train, y_train) 
x_test = lda.transform(x_test) 
```

```python
x_train # data on which model  is trained 
```

```python
x_test  # data on which moodel is tested 
```

Like PCA, you can find the variance ratios fro LDA using the explained variance ratio attribute. 

```python
# printing the variance ratios 
variance_ratios = lda.explained_variance_ratio_ 
print(variance_ratios) 
```

The above output shows that even with one component, maximum variance can be achieved. 

A Single compponent from our dataset, using the LDA. For this, we have to pass 1 as the attribute value for the n_components attribute of the LDA class. 

```python
# creating object of the LDA class 

lda = LDA(n_components = 1) 

# training the lda model on training data 
x_train = lda.fit_transform(x_train, y_train) 

x_test = lda.transform(x_test)  

variance_ratios = lda.explained_variance_ratio_ 
print(variance_ratios) 

# Here even with one feature we are getting the same results 
```

Now we will try to figure out whether or not a banknote is fake or not using a single feature. We will use the logistic regressionalgorithm for that. 

```python
# making predictions using the logistic regression 
from sklearn.linear_model import LogisticRegression 

# training the logistic regression model 
model = LogisticRegression() 
classifier = model.fit(x_train, y_train) 
classifier 
```

```python
y_pred = classifier.predict(x_test)
y_pred 
```

```python
# evauating results 
from sklearn.metrics import accuracy_score 
print(accuracy_score(y_test, y_pred))

# 98% accuracy 
```
