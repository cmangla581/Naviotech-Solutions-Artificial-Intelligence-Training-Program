                                           Data Visualization via Matplotlib
In this section, e start a formal discussion about the Matplotlib which is one of the most commonly and frequently used ython libraries for data visualization.  We can plot the graphs for the data using the pyplot. 

```python
import matplotlib.pyplot as plt 
import numpy as np 
import math 

x_vals = np.linspace(0,20,20) 
y_vals = [math.sqrt(i) for i in x_vals] 

plt.plot(x_vals, y_vals)
```

```python
import matplotlib.pyplot as plt 
import  numpy as np 
import math 

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]

fig = plt.figure() 
ax= plt.axes() 
ax.plot(x_vals, y_vals)
```

We can also increase the default size for the matplotlib plot. We can use the rcParams list of the Pyplot module and then set 2 values for the figure.figsize attribute.  The script sets the plot size to 8 inches wide and 6 inches tall. 

```python
import matplotlib.pyplot as plt 
import numpy as np 
import math 

plt.rcParams["figure.figsize"] = [8,6] 

x_vals = np.linspace(0,20,20) 
y_vals = [math.sqrt(i) for i in x_vals] 
plt.plot(x_vals, y_vals)
```

                                            Titles. Tables and Legends
We can immprove the readability and the asthetics of hte graph by adding the titles,labels and legends to the graph. To add the lebels on x and y respecrtively, we have to pass hte string values of the x label and the y label respectively. 

```python
# Example of hte code can be :   
import matplotlib.pyplot as plt 
import numpy as np 
import math  

x_vals = np.linspace(0,20,20) 
y_vals = [math.sqrt(i) for i in x_vals] 
plt.xlabel('X Values')
plt.ylabel('Y label') 
plt.title('square Roots') 
plt.plot(x_vals, y_vals)
```

In addition to changin of the titleas and the labels, we can also  specify the color of the line plot like r for red and b for blue. 

```python
import matplotlib.pyplot as plt 
import numpy as np 
import math 

x_vals = np.linspace(0,20,20) 
y_vals= [math.sqrt(i) for i in x_vals] 

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title("Square Roots") 

plt.plot(x_vals, y_vals, 'r')
```

To add a legend, we also have to pass the vlaue for the loc attribute, of the legend method of the pyplot module. In the loc attribute, you have to pass the location of your legend. 

```python
import numpy as np 
import matplotlib.pyplot as plt 
import math 

x_vals = np.linspace(0,20,20) 
y_vals = [math.sqrt(i) for i in x_vals] 

plt.xlabel("X Label") 
plt.ylabel("Y Label")
plt.title("Square Roots") 
plt.plot(x_vals, y_vals, 'r', label = "Square Root") 
plt.legend(loc = 'upper center') 
```

We can also plot the multiple line of the plots on one graph. We have to call the  plot() method twice with different values for x and y values.  The followigncode is written for such examples :   

```python
import matplotlib.pyplot as plt  
import numpy as np 
import math  

x_vals = np.linspace(0,20,20) 
y_vals = [math.sqrt(i) for i in x_vals] 

y2_vals = x_vals ** 3 

plt.xlabel("X Label") 
plt.ylabel("Y label ") 

plt.title("Square roots") 
plt.plot(x_vals, y_vals, 'r', label = 'Square roots') 
plt.plot(x_vals, y2_vals, 'b', label = 'Cube' ) 
plt.legend(loc = 'upper center')
```

                                                      Pie Charts 
Pie Charts as the name suggests dislay the percentage distribution of the values in a categorical outcome, n terms of aggregated function. 

For example in the following code, this shows the percentage distribution of jobs with respect to job categories. 
The explode parameter defines the magnitude of the split for each category in pie chart. 

```python
labels = 'IT', 'Marketing', 'Data Science', 'Finance' 
values = [500, 156,300,510] 
explode = [0.05, 0.05, 0.05, 0.05] 

plt.pie(values, explode = explode, labels= labels, autopct = '%1.1f%%', shadow = True) 
plt.show() 
```
