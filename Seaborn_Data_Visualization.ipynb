                                                     Seaborn 
we have to perform the data visualization with the Seabrn which is yet another extremely handy Python library for the data visualization.  
The Seaborn library  is used on the Matplotlib Libray. 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

plt.rcParams["figure.figsize"] = [10,8] 

tips_data = sns.load_dataset('tips') 

tips_data.head()

# rcParams refer to the Right Configuration parameters which mean that it controls the default appearance and the behaviour of all points 
# letting to globally set all the sizes and colours. 
```

                                                             Distplot 
The distplot also known as the distributional plot is used to plot a histogram of the data for a pecific column in the dataset.  

The name of the column which we want to plot is passed a s a parameter to the distplot(). 

```python
plt.rcParams["figure.figsize"] = [10,8] 
sns.distplot(tips_data["total_bill"])
```

```python
# Like the distplot, we also have a type of graph named as the displot 
sns.displot(tips_data["total_bill"], kind= 'kde')
```

The basic difference between the Displot and the Distplot that distplot is the older and the deprecated function in Seaborn for plotting the data distributions.  

While Displot is the modern figure level replacement introduced in the Seaborn offering more flexibility. 

                                                    The Joint Plot  
The Joint Plot is the plot which is used to plot the histogram distribution of the two columns, one on the x axis and the other on y axis and on the axis, and other on the y axis. A Scatter plot by efault is drawn for the points in the two columns  

To draw a joint plot, we need to call the joint plot, function.  The script plots the function for total_bill, tips columsn and the tips dataset. 

```python
sns.jointplot(x = 'total_bill', y = 'tip', data= tips_data)
```

The scatter plot can be replaced by a regression ;ine in a jointplot. We need to pass the reg as the value for the kind parameter of the jointplot function(). 

```python
sns.jointplot(x = 'size', y = 'total_bill', data = tips_data, kind = 'reg')
```

                                                          Bar Plot  
The bar plot is used to captue the relationship between the categorical and numerical column. For each unique value, in a categorical column a bar is plotted.  it dislya the numerical value of the data in a  column specified by the bar plot. 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

plt.rcParams["figure.figsize"] = [8,6] 
sns.set_style("darkgrid") 

titanic_data = sns.load_dataset('titanic') 
titanic_data.head()
```

```python
sns.barplot(x ='pclass', y = 'age', data= titanic_data)
```

                                                    Count Plot 
The count plot plots a plot like the bar plot. Unlike a bar plot, that plots the average values, the count plot display the counts of the  occurences of the records. 

The count plot function is used to plot a count plot with Seaborn. The code for the count plot for the pclass column of the Titanic Datset. 

```python
sns.countplot(x = 'pclass', data = titanic_data)
```

This shows that the  200 passengers travelled in the first class while majority of the passengers travelled in the 3rd class of Titanic ship.

                                            Pandas Bar Plot
To plot the bar plots with Pandas, we need tolist the categories and a lsit of values.  The list of categories and the list of values must have same length.  We calculate the mean of both the male and female passengers  travelling in the Titanic Ship. 

```python
titanic_data = pd.read_csv("Titanic_Dataset.csv") 
titanic_data.head() 

sex_mean = titanic_data.groupby("Sex")["Age"].mean() 

print(sex_mean) 
print(type(sex_mean.tolist()))
```

```python
df = pd.DataFrame({ 'Gender' : ["Female", "Male"], 'Age' : sex_mean.tolist() }) 

ax = df.plot.bar(x = "Gender", y="Age", figsize = (8,6))
```
