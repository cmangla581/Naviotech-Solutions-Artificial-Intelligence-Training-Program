                                                    K Means Clustering

K Means Clustering is one of the most commonly used algorithms for clustering and unlabelled data. In K Means Clustering, k refers to the number of clusters that we want in our data to be grouped into.  

In K Means Clusteering, the number of clusters has to be defined before K Clustering can be applied to the data points. 

Steps for K Means Clustering : 
1. Randomly assign the centroid values for each cluster. 
2. can be applied to large datasets
3. Scales well to unseen data sets 
4. Generalize well to clusters of various sizes and shapes 


Disadvantages of k means Clustering algorithm : 
1. The value of K has to be chosen manually. 
2. Convergence or training time depends on the initial value of k. 
3. Clustering Performance is greatly affected by the outliers. 


                                           Clustering Dummy Data with Sklearn

```python
import numpy as np 
import pandas as pd 
from sklearn.datasets import make_blobs  
from sklearn.cluster import KMeans  
import matplotlib.pyplot as plt 
```

We have to create a dummy dataset containing 500 records and 4 cluster centers. The average deviation of the record is 2. The code for generating a dummy dataset and plot data points on a plot is as : 

```python
# generating the dummy data for 500 data points with 4 clusters : 

features, labels = make_blobs(n_samples = 500, centers = 4, cluster_std = 2.00) 
features, labels 
# Printing of the whole dataset
```

```python
plt.scatter(features[:,0], features[:,1])  
# Here as we can see that 500 data points have been seperated into 4 clusters
```

```python
# performing the kMeans Clustering usinf the KMeans Class  
km_model = KMeans(n_clusters = 4) 
km_model.fit(features) 
```

Once the model is trained, we can print the cluster centersusing the cluster_centers_ attribute of the KMeans class object. 

```python
# printing the centroid values '
print(km_model.cluster_centers_)
```

In addition to finding the cluster centers, the k Means also assign a clusters label to each data point. The cluster labels are numbrs that basically serve as the cluster id.  

For example. in case of cluster ids, the ids are 0,1,2,3. 

```python
# printing the prediicted label values 
print(km_model.labels_)
```

The following code prints the clusters in differetn clors along with the cluster centers as the black data points. 

```python
# print the data points 
plt.scatter(features[:,0],  features[:,1], c = km_model.labels_, cmap = 'rainbow') 

# printing the centroids 
# printi g the centroids 
plt.scatter(km_model.cluster_centers_[:,0], km_model.cluster_centers_[:,1], s=100, c='black') 
```

```python
# The code for printing the actual 4 datasets 
plt.scatter(features[:,0], features[:,1], c=labels, cmap ='rainbow') 
```

The output shows that in the actual dataset, the clusters represented by red and yellow data points overlap. The predicted clusters do not contain any overlapping ata points.

Now, to plot the 4 clusters found by the K Means Algorithm in the iris dataset alongwith the predicted cluster centroids, the following script has to be executed : 

```python
# Printing the data points 
plt.scatter(features[:,0], features[:,1], c = km_model.labels_, cmap ='rainbow') 

# print the centroids  
plt.scatter(km_model.cluster_centers_[:,0], km_model.cluster_centers_[:,1], s =100, c = 'black') # s stands for size  
```

Till now we have been randomly initializing the value of K or the number of clusters. However, there is a way to find the ideal number of clusters.  This method is known as the elbow method, the value of the inertia obtained by training  K Means clusters with different numbers of k is plotted. 

the inertia represents the total distance between the data points within the cluster.  Small inertia means thet the predicted clusters are robust and close to the actual clusters.  

To calculate the inertia value, we can use the inerti_attribute, of the kmeans class object. the script creates the inertial values for K = 1 to 10 and the plots in the form of a line plot. 

```python
# training the KMeans on K values from 1 to 10 
loss = [] 
for i in range(1,11): 
    km = KMeans(n_clusters =i).fit(features) 
    loss.append(km.inertia_) 

# prinitng the loss against the number of clusters 
import matplotlib.pyplot as plt 
plt.plot(range(1,11), loss) 
plt.title("Finding the Optimal clusters via Elbow Method") 
plt.xlabel("Number of Clusters") 
plt.ylabel('loss') 
plt.show() 

# We can see that the value of inertia didn't decrease much after 3 clusters. 
```

```python
# training the data with 3 clusters 
km_model = KMeans(n_clusters = 3) 
km_model.fit(features) 
```

```python
# print the data points with predicted labels 
plt.scatter(features[:,0], features[:,1], c= km_model.labels_, cmap= 'rainbow') 

# Printing the predicted centroids 
plt.scatter(km_model.cluster_centers_[:,0], km_model.cluster_centers_[:,1], s =100, c = 'black') 
```

When K is 3, the above graph is the graph for the number of clusters predicted by the K Means clustering algorithm. 

These were the predicted clusters. 

Now, we plot the actual clusters and see thathow clse the actual clusters are to the predicted cluters. 

```python
from sklearn import preprocessing  
le = preprocessing.LabelEncoder()
labels = le.fit_transform(labels) 

# Print the data points with original labels 
plt.scatter(features[:,0], features[:,1], c = labels, cmap = 'rainbow') 
```

                                                 Hierarchial Clustering

Like K Means Clustering, hierarchial cluatering is another commonly used unsupervised machine learning technique for the data clustering. Hierarchial clustering can be broadly didvided into the 2 types : agglomerative clustering and divisive clustering.

Agglomerative Clustering : It  follows a bottom up approach where the individual data points are clustered together to form multiple small clusters leading to big cluster which can then be divided into small clusters using the dendograms.  

On the other hand, in the case of divisive clustering, you have one big cluster which we made into N number of small clusters. 

Steps of Hierarchial Agglomerative Clustering : 

The steps required to perform the agglomerative clustering  are as ; 
1. Consider each data point in the dataset as one cluster. Hence the number of clusters is equal to number oof data points. 
2. join the two closest data points to form a cluster. 
3. For more clusters by joining the closest clusters. Repeat the process until the one big cluster is formed. 
4. Use dendograms to divide the one big cluster into multiple small clusters. 

Benefits of Hierarchial Clusterinng : 
1. Unlike k Menas Clustering, for the hierarchial clustering, you do not have to specify the number of centroids. 
2. With dendograms, it is easire to interpret how the data has been clustered. 

Disadvantages of  Hierarchial Clustering : 
1. Doesn't scale wel; on unseen data. 
2. Has higher time compplexity compared to k maens clustering. 
3. Difficult to determine the number of clusters in case of large datsets. 

                                            Clustering Dummy Data 
                            First, we will perform the hierarchial clustering on the dummy data. 

First we perform agglomerative data clustering on 10 2-d data points only. 

```python
import pandas as pd 
import numpy as np 
from sklearn.datasets import make_blobs 
import matplotlib.pyplot as plt 
```

```python
# generating dummy data of 10 records with 2 clusters 
features, labels = make_blobs(n_samples  =10, centers = 2, cluster_std = 2.00) 
features, labels 
```

```python
# plotting the dummy data 
plt.scatter(features[:,0], features[:,1], color = 'r') 
```

```python
# adding nubers to data points 
annots = range(1, 11)

for label, x, y in zip(annots, features[:, 0], features[:, 1]):
    plt.annotate(
        label,
        xy=(x, y),
        xytext=(-3, 3),
        textcoords='offset points',
        ha='right',
        va='bottom'
    )

plt.show() 
```

Now, we plot the dendograms for the above 10 data points. To plot the dendograms, we use the dendograms and the lnkage classes from the scipy cluster. The features are passed into the linkage class. 

```python
from scipy.cluster.hierarchy import dendrogram, linkage 

dendos = linkage(features, 'single') 

annots = range(1,11) 

dendrogram(dendos, orientation = 'top', labels = annots, distance_sort ='descending', show_leaf_counts =True) 

plt.show()
```

Here, it can be seen that the points 1 and 5 are closest to each other. Hence, a cluster is formed by connecting these points. The cluster of 1 and 5 is closest to data point 10., resulting in a cluster containing the data points 1,5 and 10 in the same way.  

In the same way, the remaining clusters are formed until a big cluster is formed. 

After a big cluster is formed, select the  longest vertical line. then draw a horizontal line through it. The nuber of cllusters formed is equal to the number of vertical lines  this newly created horizontal line passes.  

Like in the above figure, 2 clusters are formed. 

Applying the agglomerative clustering to our dummy dataset as : 

```python
from sklearn.cluster import AgglomerativeClustering 

# training the agglomerative clustering model 
hc_model = AgglomerativeClustering(n_clusters = 2, metric = 'euclidean', linkage= 'ward') 

hc_model.fit_predict(features) 

# affinity is replaced by metric 
```

```python
# Plotting of the predicted clusters as: 

plt.scatter(features[:,0], features[:,1], c = hc_model.labels_ , cmap = 'rainbow') 
```

You can create the dendrograms using the feature set using the shc module from the scipy.cluster.hierarchy library.  
You have to pass the feature set to the linkage class of the shc module and then the object of the linkage class is passed to the dendrogram class. 

```python
import scipy.cluster.hierarchy as shc 

plt.figure(figsize = (10,7)) 
plt.title("Iris Dendograms") 
dend = shc.dendrogram(shc.linkage(features, method = 'ward'))  
```

The line by line explanantion of the above code can be : 

1. importing the hierarchy mmodule : It imports the hierarchial cluatering tools from SciPy. shc is just an alias to make the code better.  The module is used to create the linkage metrices and dendrograms. 

2. plt.figure(figsize = (10,7)) 
It creates a new matplotlib figure figsize = (10,7) means that width = 10 and length = 7.  It helps to make dendrogram readable especilly with many samples. 

3. Creating the dendrograms : shc.linkage(features, method = 'ward') it computes the hierarchial clustering linkage matrix. the output is a linkage matrix that describes, how the clusters are merged.
